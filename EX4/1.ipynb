```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


image = cv2.imread(r"F:\CV\build.jpg",cv2.IMREAD_GRAYSCALE)

plt.figure(figsize=(15,15))
plt.subplot(2, 2, 1)

plt.imshow(image, cmap='gray')
plt.title('Dark Image')
plt.axis('off')
hist = cv2.calcHist([image], [0], None, [256], [0, 256])
    
    # Plot histogram
plt.subplot(2, 2, 2)
plt.plot(hist, color='black')
plt.xlim([0, 256])
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')

plt.subplot(2,2,3)
contrast_image = cv2.convertScaleAbs(image,alpha=1.5,beta=50)
plt.imshow(contrast_image, cmap='gray')
plt.title('Dark contrast_Image')
plt.axis('off')
hist = cv2.calcHist([contrast_image], [0], None, [256], [0, 256])
    
    # Plot histogram
plt.subplot(2, 2, 4)
plt.plot(hist, color='black')
plt.xlim([0, 256])
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
```
