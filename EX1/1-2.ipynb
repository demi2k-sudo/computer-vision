```python
import numpy as np
import matplotlib.pyplot as plt

# Create an 8x8 matrix with random 0s and 1s
matrix = np.random.randint(0, 2, size=(8, 8))

# Display the matrix as an image
plt.imshow(matrix, cmap='gray', interpolation='none')
plt.colorbar()  # Optional: to show the color scale
plt.title('8x8 Matrix Image')
plt.show()
```
